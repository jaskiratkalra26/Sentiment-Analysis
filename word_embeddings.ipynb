```python
import pandas as pd
df = pd.read_csv('IMDB Dataset.csv')
df.head(5)
len(df)
```

```python
from transformers import BertTokenizer, TFBertModel
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = TFBertModel.from_pretrained('bert-base-uncased')
```

```python
df['sentiment'].describe()
```

```python
def sentence_embeddings(text):
  processed = tokenizer(text, return_tensors='tf',padding=True,truncation=True)
  build_model = model(**processed)
  out = build_model['last_hidden_state'][:,1,:]
  return out
```

```python
review_embeddings = []
label = []
for i in range(len(df)):
  review_embeddings.append(sentence_embeddings(df['review'][i]))
  label.append(df['sentiment'][i])
```

```python
review_embeddings[0]
```

```python
data = {'embeddings':review_embeddings,
        'labels':label}
```

```python
import pickle
with open('review_embeddings.pkl','wb') as f:
  pickle.dump(data,f)
```
